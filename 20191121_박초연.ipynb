	4.	team 승수. 팀명. salary 연봉. 팀명 2016년  groupby(팀명) 30개 팀 평균 데이터를 구하시오.   그리고 1승을 하기 위하여 연봉을 얼마나 더 지급해야 하나? optional.

```
import pandas as pd
df = pd.read_csv('http://wolfpack.hnu.ac.kr/Stat_Notes/example_data/baseball.csv')
```

```
ct = pd.DataFrame(df.Position.value_counts())
ct.reset_index(inplace=True)
ct.columns=['Position', 'count']
df0 = pd.merge(df,ct,on='Position',how='inner')
#df1 = df0[df0['count']>=16]
df1 = df0[df0.Salary!='.']
df1['Salary']=pd.to_numeric(df1.Salary)
```

```
df2=df1[df1.Position == '1B']
```

```
hit = df2['Hits']/df2['TimesatBat']
```

```
df2["hit"] = hit
```

## 1.	1루 타자가 타율을 1푼 올리면 연봉이 얼마나 올라가나? 

```
import statsmodels.api as sm
y=df2['Salary']
X=sm.add_constant(df2['hit'])
model=sm.OLS(y, X).fit()
model.summary()
```

타율을 1푼 올리면 연봉이 92.280140천불만큼 올라간다.

##	2.	지명타자가 타율을 1푼 올리면 연봉이 얼마나 올라가나?

```
df3=df1[df1.Position == '3B']
```

```
hit0 = df3['Hits']/df3['TimesatBat']
df3["hit"] = hit0
```

```
import statsmodels.api as sm
y=df3['Salary']
X=sm.add_constant(df3['hit'])
model=sm.OLS(y, X).fit()
model.summary()
```

타율을 1푼 올리면 연봉이 94.119454천불만큼 올라간다.
