#문제4.
선수연봉 95% 신뢰구간을 구하시오.

```
from scipy.stats import t
import numpy as np
def CI(alpha,df):
  xbar=df.mean()
  sd=df.std()
  n=df.shape[0]
  lb=xbar-t.ppf(1-alpha,df=n-1)*sd/np.sqrt(n)
  ub=xbar+t.ppf(1-alpha,df=n-1)*sd/np.sqrt(n)
  print('모평균 추정치=%.3f, 모평균 %2d%% 신뢰구간 : 하한=%.3f, 상한=%.3f' % (xbar,(1-alpha)*100,lb,ub))
```
