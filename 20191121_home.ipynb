#### team 승수. 팀명. salary 연봉. 팀명 2016년 groupby(팀명) 30개 팀 평균 데이터를 구하시오.

```
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```
team=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/Teams.csv')
salary=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/Salaries.csv')
```

```
team = team[team.yearID == 2016]
team.groupby(by='teamID')
team = team[['teamID','W']]
team = team.replace('.',np.nan)
```

```
salary = salary[salary.yearID == 2016]
salary = salary[['salary','teamID']]
salary = salary.groupby(by='teamID').mean()
salary = salary.replace('.',np.nan)
```

```
t_s = pd.merge(team,salary,on='teamID',how='inner')
t_s = t_s.replace('.',np.nan)
```

```
t_s
```

#### 그리고 1승을 하기 위하여 연봉을 얼마나 더 지급해야 하나? optional.

```
import statsmodels.api as sm
y=t_s['salary']
X=sm.add_constant(t_s['W'])
model=sm.OLS(y, X).fit()
model.summary()
```

1승을 하기 위해서는 887.6천불 더 지급해야하며,

이 결과는 Prob (F-statistic)가 0.001보다 작으므로 유의하다.

R-squared가 0.374인 것을 보아 연봉은 승수에 37.4% 의 영향을 미치고,

나머지(62.6%)는 승수 외의 다른 요인으로 인해 변동된다.
