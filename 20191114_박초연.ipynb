```
import mpl_finance
from mpl_finance import candlestick2_ohlc
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.ticker as ticker
import numpy as np
```

```
fig = plt.figure(figsize=(8,6))
ax_main = plt.subplot2grid((5, 1), (0, 0), rowspan=3)
ax_sub = plt.subplot2grid((5, 1), (3, 0))

ax_main.set_title('celltrion 2019~today time-plot',fontsize=15)
mpl_finance.candlestick2_ohlc(ax_main, 
                              df['Open'], df['High'], df['Low'], df['Close'],
                              width=0.5, colorup='r', colordown='b')

ax_sub.bar(df.index,df['Volume'])

plt.tight_layout()
plt.show()
```

```
import plotly.graph_objects as go
import pandas as pd

fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'], high=df['High'],
                low=df['Low'], close=df['Close'])
                      ])

fig.update_layout(
    title=' ',
    yaxis_title=' ',
    shapes = [dict(
        x0='2019-01-01', x1='2019-06-30', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2019-01-01', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='left', text='Increase Period Begins')])
fig.show()
```
