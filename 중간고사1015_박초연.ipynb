## 선수정보, 연봉, 타자, 투수, 팀 데이터 읽어오기

```
import pandas as pd
player=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/People.csv')
salary=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/Salaries.csv')
batter=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/Batting.csv')
pitcher=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/Pitching.csv')
team=pd.read_csv('/content/drive/My Drive/통계적방법19_박초연/Teams.csv')
```

## 만든 데이터프레임 변수명 출력

```
print('선수정보',player.columns,'\n연봉',salary.columns,
'\n타자',batter.columns,'\n투수',pitcher.columns,'\n팀',team.columns)
```

## 만든 데이터프레임 크기 보기

```
print('선수정보',player.shape,'\n연봉',salary.shape,'\n타자',batter.shape,
'\n투수',pitcher.shape,'\n팀',team.shape)
```

## 선수정보 + 연봉정보 합치기 (선수_연봉)
합친 데이터프레임 이름 : (선수_연봉) ps

```
ps=pd.merge(salary,player,on='playerID',how='inner')
ps.sort_values(['yearID','playerID'],inplace=True)
print(ps.shape,'\n',ps.columns)
```

## 타자 정보 + 선수_연봉정보 합치기

```
bat=pd.merge(ps,batter,on=['yearID','playerID'],how='inner')
print(bat.shape,'\n',bat.columns)
```

##문제1. 2016년 연봉평균은 2015년 연봉평균보다 올랐는가?
(이상치 제거X, 동질성 검정O)

>귀무가설(H0) : 2016년 연봉 평균(Mu1)은 2015년 연봉 평균(Mu2)과 같다. Mu1=Mu2

>대립가설(H1) : 2016년 연봉 평균(Mu1)은 2015년 연봉 평균(Mu2)보다 높다. Mu1hat>Mu2hat

유의수준:0.05

```
salary15=salary[(salary.yearID==2015)]
salary16=salary[(salary.yearID==2016)]
```

```
ms15=salary15.salary.mean()
ms16=salary16.salary.mean()
print(ms15,'\n',ms16)
```

분산 동질성 검정

>귀무가설(H0) : 2016년 연봉분산(V(X1))은 2015년 연봉분산(V(X2))과 같다. Mu1=Mu2 V(X1)=V(X2)

>대립가설(H1):2016년 연봉 분산(V(X1))은 2015년 연봉 분산V(X2)과 다르다. V(X1)!=V(X2)


-분산이 같은경우(귀무가설채택=pvalue가 크다) : stats.ttest_ind(data1,data2,equal_var=True)

-분산이 다를경우(귀무가설기각=pvalue가 작다) : stats.ttest_ind(data1,data2,equal_var=False)

(pvalue가 크다 작다는 유의수준으로 판단, 언급없으면 0.05)

```
import scipy.stats
scipy.stats.levene(salary16.salary,salary15.salary)
```

>결론 : pvalue가 0.48로 유의수준 0.05보다 큰값을 보여, 귀무가설 채택
즉, 두 집단의 모분산은 같다.

```
import scipy.stats
scipy.stats.ttest_ind(salary16.salary,salary15.salary,equal_var=True)
```

결론 : 분산이 같을 때, 두 평균이 같을 확률은 0.73으로 매우 크며 유의수준 0.05보다 커, 0.75의 높은 확률로 관측되었기 때문에 H0 연봉 평균이 같다는 가설이 맞다고 판단하여 귀무가설을 채택하며, 연봉 평균은 0.34일 것이다.
